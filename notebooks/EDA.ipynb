## Introduction
In this notebook, we will explore the basics of Exploratory Data Analysis (EDA) for medical image analysis using Machine Learning techniques.

## What is Exploratory data analysis (EDA)?
#### EDA is a crucial step in data analysis where statistical techniques and visualizations are used to uncover patterns, trends, and relationships in the data. It helps in understanding the data's structure, identifying outliers, missing values, and formulating hypotheses for further investigation. EDA aids in making informed decisions, selecting appropriate modeling techniques, and guiding the overall data analysis process. 
---

## Step 1: Importing Required Libraries
We begin by importing the necessary libraries for our analysis. We'll use Python libraries like NumPy, Pandas, and Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import glob
```

## Step 2: Loading the Dataset
Next, we load the medical image dataset that will be used for our analysis. The dataset may consist of X-ray, MRI, or other medical images. 
In our case we have X-rays.

```python
Normal = glob.glob('../Data/chest_xray/train/NORMAL/*.jpeg') # This will load all the .jpeg files from the NORMAL examples from the train samples directory of the dataset.
Pneumonia = glob.glob('../Data/chest_xray/train/PNEUMONIA/*.jpeg') # This will load all the .jpeg files from the Pneumonia examples from the train samples directory of the dataset.
```

## Step 3: Exploring the Dataset
We start by examining the structure and characteristics of the dataset. We can check the dimensions of the images, the number of classes, and other relevant information.

```python
print(f"Normal Image path sample: {Normal[0:10]}\n\n\n")
print(f"Pneumonia Image path sample: {Pneumonia[0:10]}")
```

```python
print(f"Total Normal samples: {len(Normal)}")
print(f"Total Pneumonia samples: {len(Pneumonia)}")
```

### So as we can see, the dataset that we have is an unbalanced dataset (we'll get back to this while training the model)

# Step 4: Visualising the data

```python
figure = plt.figure(figsize=(25, 10))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = np.random.randint(len(Normal), size=(1,)).item()
    img_path = Normal[sample_idx]
    img = Image.open(img_path).resize((512, 512))
    img = np.array(img)
    figure.add_subplot(rows, cols, i)
    
    label = img_path.split('/')[-2]
    plt.title(label)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
plt.show()
```

```python
figure = plt.figure(figsize=(25, 10))
cols, rows = 3, 3
for i in range(1, cols * rows + 1):
    sample_idx = np.random.randint(len(Pneumonia), size=(1,)).item()
    img_path = Pneumonia[sample_idx]
    img = Image.open(img_path).resize((512, 512))
    img = np.array(img)
    figure.add_subplot(rows, cols, i)
    
    label = img_path.split('/')[-2]
    plt.title(label)
    plt.axis("off")
    plt.imshow(img, cmap="gray")
plt.show()
```

```python
plt.pie(x = (len(Normal), len(Pneumonia)), labels=[f'Normal: {len(Normal)}', f'Pneumonia: {len(Pneumonia)}'])
plt.title("Data Distribution")
plt.show()
```
